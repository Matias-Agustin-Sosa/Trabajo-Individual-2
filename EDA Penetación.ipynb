# EDA
Importamos las librerias necesarias para leer datasets y graficar:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Penetración-hogares

```python
df_p = pd.read_csv(r"DataSets\Penetracion-hogares.csv", sep = ";")
```

Vemos si hay valores faltantes en este dataset:

```python
df_p.isnull().sum()
```

No hay valores faltantes (Nulos), pasamos a revisar filas duplicadas:

```python
df_p[df_p.duplicated(keep = False)]
```

Busco outliers en las columnas Accesos por cada 100 hogares, mediante Gráficos de dispersión:

```python
plt.figure(figsize = (7,5))
plt.plot(df_p["Accesos por cada 100 hogares"], "bo")
plt.title("Accesos por cada 100 hogares")
plt.show()
```

En este caso el gráfico no parece aportar mucha información.

```python
df_p[["Año","Trimestre", "Provincia","Accesos por cada 100 hogares"]].sort_values(by = "Accesos por cada 100 hogares",ascending = False).head(15)
```

Se puede notar como en Capital Federal hay una moyor presencia del servicio.

```python
df_p[["Año","Trimestre", "Provincia","Accesos por cada 100 hogares"]].sort_values(by = "Accesos por cada 100 hogares",ascending = True).head(15)
```

Caso inverso, San Luis es la provincia con menor presencia del servicio.

## Penetración-poblacion

```python
df = pd.read_csv(r"DataSets\Penetración-poblacion.csv", sep = ";")
```

Vemos si hay valores faltantes en este dataset:

```python
df.isnull().sum()
```

No hay valores faltantes (Nulos), pasamos a revisar filas duplicadas:

```python
df[df.duplicated(keep = False)]
```

Busco outliers en las columnas Accesos por cada 100 hab, mediante Gráficos de dispersión:

```python
plt.figure(figsize = (7,5))
plt.plot(df["Accesos por cada 100 hab"], "yo")
plt.title("Accesos por cada 100 hab")
plt.show()
```

Parecen haber varios outliers.

```python
df[["Año","Trimestre", "Provincia","Accesos por cada 100 hab"]].sort_values(by = "Accesos por cada 100 hab",ascending = False).head(15)
```

Lógicamente, Capital Federal lidera, esto es coherente con el análisis de la tabla anterior. Igual se puede notar en el gráfico que la diferencia de Capital Federal con el resto es mucho mayor que en el gráfico de "Accesos por cada 100 hogares".

```python
df[["Año","Trimestre", "Provincia","Accesos por cada 100 hab"]].sort_values(by = "Accesos por cada 100 hab",ascending = True).head(15)
```

En el caso de San Luis, también era de esperarse por lo mismo que en la situación anterior.
